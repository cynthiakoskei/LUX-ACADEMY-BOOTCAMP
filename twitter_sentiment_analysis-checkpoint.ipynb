```python
#import necessary modules
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
```

```python
# Define a list of example tweets
tweets = [
    "I love data science!",
    "This is a great tutorial on sentiment analysis.",
    "I'm feeling neutral about this.",
    "I dislike negative sentiment.",
]
```

```python
# Define a function to perform sentiment analysis
def analyze_sentiment(tweet_text):
    analysis = TextBlob(tweet_text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
# Analyze the sentiment of each tweet and store the results
sentiments = [analyze_sentiment(tweet) for tweet in tweets]
```

```python
# Count the number of tweets in each sentiment category
sentiment_counts = {
    'Positive': sentiments.count('Positive'),
    'Negative': sentiments.count('Negative'),
    'Neutral': sentiments.count('Neutral')
}
```

```python
# Create a bar chart to visualize sentiment distribution
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.barplot(x=list(sentiment_counts.keys()), y=list(sentiment_counts.values()))
plt.title('Sentiment Analysis of Example Tweets')
plt.xlabel('Sentiment')
plt.ylabel('Count')

# Show the bar chart
plt.show()
```
